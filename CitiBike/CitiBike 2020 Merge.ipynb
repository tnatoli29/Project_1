```python
#Dependences 
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# Take in all of our CitiBike data and read it into pandas
jan_2020 = "../Resources/JC-202001-citibike-tripdata.csv"
feb_2020 = "../Resources/JC-202002-citibike-tripdata.csv"
mar_2020 = "../Resources/JC-202003-citibike-tripdata.csv"
apr_2020 = "../Resources/JC-202004-citibike-tripdata.csv"
may_2020 = "../Resources/JC-202005-citibike-tripdata.csv"
jun_2020 = "../Resources/JC-202006-citibike-tripdata.csv"
jul_2020 = "../Resources/JC-202007-citibike-tripdata.csv"
aug_2020 = "../Resources/JC-202008-citibike-tripdata 2.csv"

jan_2020_df = pd.read_csv(jan_2020)
feb_2020_df = pd.read_csv(feb_2020)
mar_2020_df = pd.read_csv(mar_2020)
apr_2020_df = pd.read_csv(apr_2020)
may_2020_df = pd.read_csv(may_2020)
jun_2020_df = pd.read_csv(jun_2020)
jul_2020_df = pd.read_csv(jul_2020)
aug_2020_df = pd.read_csv(aug_2020)
jul_2020_df
```

```python
# combine into one DF
combined2020_CitiBike_df = pd.merge(jan_2020_df, feb_2020_df, how='outer')
combined2020_CitiBike_df
combined2020_CitiBike_df3 = pd.merge(combined2020_CitiBike_df, mar_2020_df, how='outer')
combined2020_CitiBike_df3
combined2020_CitiBike_df4 = pd.merge(combined2020_CitiBike_df3, apr_2020_df, how='outer')
combined2020_CitiBike_df4
combined2020_CitiBike_df5 = pd.merge(combined2020_CitiBike_df4, may_2020_df, how='outer')
combined2020_CitiBike_df5
combined2020_CitiBike_df6 = pd.merge(combined2020_CitiBike_df5, jun_2020_df, how='outer')
combined2020_CitiBike_df6
combined2020_CitiBike_df7 = pd.merge(combined2020_CitiBike_df6, jul_2020_df, how='outer')
combined2020_CitiBike_df7
combined2020_CitiBike_df8 = pd.merge(combined2020_CitiBike_df7, aug_2020_df, how='outer')
combined2020_CitiBike_df8
```

```python
combined2020_CitiBike_df8["gender"] = combined2020_CitiBike_df8["gender"].map(str)
combined2020_CitiBike_df8["gender"] = combined2020_CitiBike_df8["gender"].replace(['0','1','2'],['Unknown','Male','Female'])
combined2020_CitiBike_df8
combined2020_CitiBike_df8 = combined2020_CitiBike_df8.rename(columns = {"tripduration": "tripduration (seconds)"})
combined2020_CitiBike_df8
```

```python
#splitting start time column
combined2020_CitiBike_df8['starttime'] = pd.to_datetime(combined2020_CitiBike_df8['starttime'])
combined2020_CitiBike_df8['start date'] = combined2020_CitiBike_df8['starttime'].dt.date
combined2020_CitiBike_df8['start time'] = combined2020_CitiBike_df8['starttime'].dt.time
combined2020_CitiBike_df8.head()

#splitting stop time column
combined2020_CitiBike_df8['stoptime'] = pd.to_datetime(combined2020_CitiBike_df8['stoptime'])
combined2020_CitiBike_df8['end date'] = combined2020_CitiBike_df8['stoptime'].dt.date
combined2020_CitiBike_df8['end time'] = combined2020_CitiBike_df8['stoptime'].dt.time
combined2020_CitiBike_df8.head()
```

```python
# Display the total number of rides for 2020
count = len(combined2020_CitiBike_df8["tripduration (seconds)"])
print(f"There were a total of {count} between January 1, 2020 through August 31,2020")
```

```python
# Create a pie chart based upon use of service by gender
bike_pie = gender_type.plot(kind="pie", y=gender_type, title=("2020 Bike Service by Gender"))
bike_pie.set_ylabel("Gender")
plt.tight_layout()
plt.axis("equal")
plt.show()
```

```python
# average trip duration broken down by month

jancount2020 = len(jan_2020_df["tripduration"])
jantripduration2020 = jan_2020_df["tripduration"].sum()
janaveragetripduration2020 = jantripduration2020/jancount2020
janminutes = janaveragetripduration2020/60
janminutes.round(2)
print(f"The average trip duration for January 2020 was {janminutes.round(2)}")

febcount2020 = len(feb_2020_df["tripduration"])
febtripduration2020 = feb_2020_df["tripduration"].sum()
febaveragetripduration2020 = febtripduration2020/febcount2020
febminutes = febaveragetripduration2020/60
febminutes.round(2)
print(f"The average trip duration for Febuary 2020 was {febminutes.round(2)}")

marcount2020 = len(mar_2020_df["tripduration"])
martripduration2020 = mar_2020_df["tripduration"].sum()
maraveragetripduration2020 = martripduration2020/marcount2020
marminutes = maraveragetripduration2020/60
marminutes.round(2)
print(f"The average trip duration for March 2020 was {marminutes.round(2)}")

aprcount2020 = len(apr_2020_df["tripduration"])
aprtripduration2020 = apr_2020_df["tripduration"].sum()
apraveragetripduration2020 = aprtripduration2020/aprcount2020
aprminutes = apraveragetripduration2020/60
aprminutes.round(2)
print(f"The average trip duration for April 2020 was {aprminutes.round(2)}")

maycount2020 = len(may_2020_df["tripduration"])
maytripduration2020 = may_2020_df["tripduration"].sum()
mayaveragetripduration2020 = maytripduration2020/maycount2020
mayminutes = mayaveragetripduration2020/60
mayminutes.round(2)
print(f"The average trip duration for May 2020 was {mayminutes.round(2)}")

juncount2020 = len(jun_2020_df["tripduration"])
juntripduration2020 = jun_2020_df["tripduration"].sum()
junaveragetripduration2020 = juntripduration2020/juncount2020
junminutes = junaveragetripduration2020/60
junminutes.round(2)
print(f"The average trip duration for June 2020 was {junminutes.round(2)}")

julcount2020 = len(jul_2020_df["tripduration"])
jultripduration2020 = jul_2020_df["tripduration"].sum()
julaveragetripduration2020 = jultripduration2020/julcount2020
julminutes = julaveragetripduration2020/60
julminutes.round(2)
print(f"The average trip duration for July 2020 was {julminutes.round(2)}")

augcount2020 = len(aug_2020_df["tripduration"])
augtripduration2020 = aug_2020_df["tripduration"].sum()
augaveragetripduration2020 = augtripduration2020/augcount2020
augminutes = augaveragetripduration2020/60
augminutes.round(2)
print(f"The average trip duration for August 2020 was {augminutes.round(2)}")

months = {"Months": ["January,2020","February,2020","March,2020","April,2020","May,2020","June,2020","July,2020","August,2020"],
          "Avg trip duration in minutes" : [janminutes.round(2), febminutes.round(2), marminutes.round(2),
                                            aprminutes.round(2),mayminutes.round(2), junminutes.round(2),
                                            julminutes.round(2), augminutes.round(2)]}
Avg_tripduration = pd.DataFrame(months, columns = ["Months", "Avg trip duration in minutes"])
Avg_tripduration
```

```python
# plot a line graph for Average trip duration in minutes 
x_axis = Avg_tripduration["Months"]
y_axis = Avg_tripduration["Avg trip duration in minutes"]
plt.plot(x_axis, y_axis, color = "orchid")
plt.xlabel("Month")
plt.ylabel("Average trip duration in minutes")
plt.title("Average trip duration in minutes per Month")
plt.grid()
plt.show()
```

```python
# plot a bar chart for Average trip duration in minutes 

#Set the x and y axis
x_axis = Avg_tripduration["Months"] 
y_axis = Avg_tripduration["Avg trip duration in minutes"]
#Create the bar chart 
plt.bar(x_axis, y_axis, color = "blue")

#Set the labels and title
plt.title("Average trip duration in minutes per Month")
plt.xlabel("Month")
plt.ylabel("Average trip duration in minutes")
plt.grid()
plt.show()
```
