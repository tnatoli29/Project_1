```python
#Dependences
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Take in all of our CitiBike data and read it into pandas
jan_2020 = "../Resources/JC-202001-citibike-tripdata.csv"
feb_2020 = "../Resources/JC-202002-citibike-tripdata.csv"
mar_2020 = "../Resources/JC-202003-citibike-tripdata.csv"
apr_2020 = "../Resources/JC-202004-citibike-tripdata.csv"
may_2020 = "../Resources/JC-202005-citibike-tripdata.csv"
jun_2020 = "../Resources/JC-202006-citibike-tripdata.csv"
jul_2020 = "../Resources/JC-202007-citibike-tripdata.csv"
aug_2020 = "../Resources/JC-202008-citibike-tripdata 2.csv"

jan_2020_df = pd.read_csv(jan_2020)
feb_2020_df = pd.read_csv(feb_2020)
mar_2020_df = pd.read_csv(mar_2020)
apr_2020_df = pd.read_csv(apr_2020)
may_2020_df = pd.read_csv(may_2020)
jun_2020_df = pd.read_csv(jun_2020)
jul_2020_df = pd.read_csv(jul_2020)
aug_2020_df = pd.read_csv(aug_2020)
aug_2020_df.head()
```

```python
# combine into one DF
combined2020_CitiBike_df = pd.merge(jan_2020_df, feb_2020_df, how='outer')
combined2020_CitiBike_df
combined2020_CitiBike_df3 = pd.merge(combined2020_CitiBike_df, mar_2020_df, how='outer')
combined2020_CitiBike_df3
combined2020_CitiBike_df4 = pd.merge(combined2020_CitiBike_df3, apr_2020_df, how='outer')
combined2020_CitiBike_df4
combined2020_CitiBike_df5 = pd.merge(combined2020_CitiBike_df4, may_2020_df, how='outer')
combined2020_CitiBike_df5
combined2020_CitiBike_df6 = pd.merge(combined2020_CitiBike_df5, jun_2020_df, how='outer')
combined2020_CitiBike_df6
combined2020_CitiBike_df7 = pd.merge(combined2020_CitiBike_df6, jul_2020_df, how='outer')
combined2020_CitiBike_df7
combined2020_CitiBike_df8 = pd.merge(combined2020_CitiBike_df7, aug_2020_df, how='outer')
combined2020_CitiBike_df8.head()
```

```python
#Changing gender column to male, female, and unknown vs 0,1,2
combined2020_CitiBike_df8["gender"] = combined2020_CitiBike_df8["gender"].map(str)
combined2020_CitiBike_df8["gender"] = combined2020_CitiBike_df8["gender"].replace(['0','1','2'],['Unknown','Male','Female'])
combined2020_CitiBike_df8.head()
```

```python
#splitting start time column
combined2020_CitiBike_df8['starttime'] = pd.to_datetime(combined2020_CitiBike_df8['starttime'])
combined2020_CitiBike_df8['start date'] = combined2020_CitiBike_df8['starttime'].dt.date
combined2020_CitiBike_df8['start time'] = combined2020_CitiBike_df8['starttime'].dt.time
combined2020_CitiBike_df8.head()
```

```python
#splitting end time time column 
combined2020_CitiBike_df8['stoptime'] = pd.to_datetime(combined2020_CitiBike_df8['stoptime'])
combined2020_CitiBike_df8['end date'] = combined2020_CitiBike_df8['stoptime'].dt.date
combined2020_CitiBike_df8['end time'] = combined2020_CitiBike_df8['stoptime'].dt.time
combined2020_CitiBike_df8.head()
```

```python
# Create bins in which to place values based upon birth year
bins = [1900, 1946, 1964, 1979, 1994, 2015]

# Create labels for these bins
bin_labels = ["Silent Generation", "Baby Boomers", "Gen X", "Millennials", "Gen Z"]

#Create a new column for which been each row falls into
combined2020_CitiBike_df8["binned"] = pd.cut(combined2020_CitiBike_df8["birth year"], bins)

#Get the value count for each bin group
bin_count = pd.cut(combined2020_CitiBike_df8["birth year"], bins=bins).value_counts()

#Group by the binned column
bin_group = combined2020_CitiBike_df8.groupby("binned")

#Print the number of values in each bin group
print(bin_group["birth year"].count())
```

```python
#Find the most popular routes 
most_popular_routes_df = combined2020_CitiBike_df8.groupby(["start station name",
                                                            "end station name"])\
                                                            .size().reset_index(name = "Number of Trips")
most_popular_routes_df = most_popular_routes_df.sort_values(["Number of Trips"], ascending = False)
most_popular_routes_df.head()
```

```python
most_popular_routes_df["Popular Routes"] = most_popular_routes_df["start station name"].str.cat(most_popular_routes_df["end station name"],sep= " to ")
most_popular_routes_df.head()
```

```python
#Set up bar chart for most popular routes
routes_x_axis = most_popular_routes_df["Popular Routes"].head()
routes_y_axis = most_popular_routes_df["Number of Trips"].head()

plt.barh(routes_x_axis, routes_y_axis, color = "dodgerblue")
plt.title("Number or Rides From One Locations to the Next")
plt.xlabel("Number of Rides per Route")
plt.ylabel("Busiest Start to Stop Location")
plt.tight_layout()
plt.savefig("../Images/rides2020_bar")
plt.show()
```

```python
#Drop pins on the most popular start and stop locations 
import gmaps
from config import g_key

location_df = combined2020_CitiBike_df8[["start station name", "start station latitude",
                                         "start station longitude", "end station id", "end station name", 
                                         "end station latitude","end station longitude"]]

location_df.head()
```

```python
#Group the functions
pop_locations = location_df.groupby(["start station name", "start station latitude", "start station longitude",
                                     "end station name", "end station latitude", "end station longitude"])\
                                        .size().reset_index(name = "Number of Trips")
pop_locations= pop_locations.sort_values(["Number of Trips"], ascending = False)
pop_locations.head()
```

```python
#Set up varibales to for map 
start_locations = pop_locations[["start station latitude", "start station longitude"]].astype(float)
end_locations = pop_locations[["end station latitude", "end station longitude"]].astype(float)
```

```python
#Create the marker layer map 
fig = gmaps.figure(center=(48.0, 0.0), zoom_level=2)

start_marker_layer =  gmaps.marker_layer(start_locations.head(), hover_text='', label='')

fig.add_layer(start_marker_layer)

fig
```

```python
##Create the heat layer map 
fig = gmaps.figure(center=(48.0, 0.0), zoom_level=2)

end_marker_layer =  gmaps.marker_layer(end_locations.head(), hover_text='', label='')

fig.add_layer(end_marker_layer)

fig
```
